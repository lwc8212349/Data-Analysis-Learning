```python
#调整subplot周围的间距
#默认情况下，matplotlib会在subplot外围留下一定的边距，并在subplot之间留下一定的间距
#间距跟图像的高度和宽度有关。利用Figure的subplots_adjust方法可以轻而易举的修改间距
fix,axes=plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
import pandas as pd
data=pd.read_csv(r'C:\Users\Administrator\Desktop\examples\examples_data_for_data_analysis\spx.csv',
                index_col=0,parse_dates=True)
```

```python
spx=data['SPX']
```

```python
spx.plot(ax=ax,style='k--')
```

```python
crisis_data=[(datetime(2007,10,11),'Peak of bull market'),
            (datetime(2008,3,12),'Bear Stearns Fails'),
            (datetime(2008,9,15),'Lehman Bankruptcy')]
```

```python
for date,label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
               xytext=(date,spx.asof(date)+225),
                arrowprops=dict(facecolor='black',headwidth=4,
                               width=2,
                               headlength=4),
                horizontalalignment='left',verticalalignment='top')
```

```python
ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1800])
```

```python
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
#散布图或点图
macro=pd.read_csv(r'C:\Users\Administrator\Desktop\examples\examples_data_for_data_analysis/macrodata.csv')
```

```python
data=macro[['cpi','m1','tbilrate','unemp']]
```

```python
trans_data=np.log(data).diff().dropna()
```

```python
trans_data[-5:]
```

```python
sns.regplot('m1','unemp',data=trans_data)
```

```python
plt.title('Changes in log %s versus log %s'%('m1','unemp'))
```

```python
sns.pairplot(trans_data,diag_kind='kde',plot_kws={'alpha':0.2})
```

```python
sns.factorplot(x='day',y='tip_pct',hue='time',col='smoker',
              kind='bar',data=tips[tips.tip_pct<1])
```
